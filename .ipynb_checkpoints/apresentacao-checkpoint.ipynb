# Metodologia

````python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn3, venn2_circles, venn3_circles
````

# Resultados

## Diagrama geral <a id="diagrama-geral"></a>

Ler a tabela base.csv usando a função ```read_csv```:

````python
base_df = pd.read_csv('base.csv', sep=',', index_col='ID')
````

Separar todos os subconjuntos possíveis fazendo um filtro com base nas colunas: educação, saúde e mobilidade:

````python
educacao_col = base_df['Educação']
saude_col = base_df['Saúde']
mobilidade_col = base_df['Mobilidade']

aluno_df = base_df.loc[(educacao_col == 1) & (saude_col == 0) & (mobilidade_col == 0)]
dengue_df = base_df.loc[(educacao_col == 0) & (saude_col == 1) & (mobilidade_col == 0)]
onibus_df = base_df.loc[(educacao_col == 0) & (saude_col == 0) & (mobilidade_col == 1)]
aluno_dengue_df = base_df.loc[(educacao_col == 1) & (saude_col == 1) & (mobilidade_col == 0)]
dengue_onibus_df = base_df.loc[(educacao_col == 0) & (saude_col == 1) & (mobilidade_col == 1)]
aluno_onibus_df = base_df.loc[(educacao_col == 1) & (saude_col == 0) & (mobilidade_col == 1)]
aluno_dengue_onibus_df = base_df.loc[(educacao_col == 1) & (saude_col == 1) & (mobilidade_col == 1)]
````

Desenhar o diagrama usando a função ```venn3``` e o módulo pyplot da biblioteca matplotlib para gerar a imagem:

````python
subsets = (
    aluno_df.shape[0], 
    dengue_df.shape[0], 
    aluno_dengue_df.shape[0], 
    onibus_df.shape[0], 
    aluno_onibus_df.shape[0], 
    dengue_onibus_df.shape[0], 
    aluno_dengue_onibus_df.shape[0]
)

plt.figure(figsize=(10, 10))
venn3(subsets, set_labels=('Aluno', 'Dengue', 'Ônibus'), set_colors=('red', 'green', 'blue'))
venn3_circles(subsets)
plt.show()
````

 ## 1) Relatório Educação
 
 Informar nome, data de nascimento e id dos cidadãos de XPTO que frequentaram a escola, menos os cidadãos que tiveram dengue.

### Tabela

````python
aluno_df = base_df.loc[(educacao_col == 1) & (saude_col == 0)]
dengue_df = base_df.loc[(saude_col == 1) & (educacao_col == 0)]
aluno_dengue_onibus_df = base_df.loc[(educacao_col == 1) & (saude_col == 1)]
table_df = aluno_df[['Nome', 'Data de Nascimento']]

table_df.to_csv('relatorios/relatorio-educacao.csv')
````

### Diagrama

````python
subsets = (aluno_df.shape[0], dengue_df.shape[0], aluno_dengue_onibus_df.shape[0])
v = venn2(subsets, set_labels=('Aluno', 'Dengue'))

for area in ['10', '01', '11']:
    v.get_patch_by_id(area).set_color('blue' if area == '10' else 'white')

venn2_circles(subsets)
plt.show()
````

## 2) Relatório Saúde

Informar nome, data de nascimento e data que tiveram dengue dos cidadãos de XPTO que frequentaram o posto de saúde, menos os cidadãos que não utilizam ônibus.

### Tabela

````python
dengue_df = base_df.loc[(saude_col == 1) & (mobilidade_col == 0)]
onibus_df = base_df.loc[(mobilidade_col == 1) & (saude_col == 0)]
dengue_onibus_df = base_df.loc[(saude_col == 1) & (mobilidade_col == 1)]
table_df = dengue_onibus_df[['Nome', 'Data de Nascimento', 'Data da Dengue']]

table_df.to_csv('relatorios/relatorio-saude.csv')
````

### Diagrama

````python
subsets = (dengue_df.shape[0], onibus_df.shape[0], dengue_onibus_df.shape[0])
v = venn2(subsets, set_labels=('Dengue', 'Ônibus'))

for area in ['10', '01', '11']:
    v.get_patch_by_id(area).set_color('blue' if area == '11' else 'white')
    
venn2_circles(subsets)
plt.show()
````
